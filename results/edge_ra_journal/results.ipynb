```python
import pandas as pd

df_noext = pd.read_csv(f"results_raandext_noext.csv")
df_noext.columns = df_noext.columns.str.strip()
df_noext = df_noext[["Testcase","Total cost","Avg. maximum unattested time","Avg. maximum unattested time %" ,"Avg. time spent attesting" ,"Avg. time spent attesting %" ,"Avg. worst case response time" ,"Avg. avg response time" ,"Avg. avg response time %"]]

df_ext = pd.read_csv(f"results_raandext.csv")
df_ext.columns = df_ext.columns.str.strip()
df_ext = df_ext[["Testcase","Total cost","Avg. maximum unattested time","Avg. maximum unattested time %" ,"Avg. time spent attesting" ,"Avg. time spent attesting %" ,"Avg. worst case response time" ,"Avg. avg response time" ,"Avg. avg response time %"]]
```

```python
sum_perc_dec = 0
res1 = pd.DataFrame(columns=["", "NOEXT", "EXT"])
print("Average maximum unattested time")
for i in range(11):
    tc1 = df_noext.iloc[i]["Testcase"]
    tc2 = df_ext.iloc[i]["Testcase"]
    assert tc1 == tc2
    amut_noext = df_noext.iloc[i]["Avg. maximum unattested time"].item()
    amut_ext = df_ext.iloc[i]["Avg. maximum unattested time"].item()
    
    perc_decrease = ((float(amut_noext) - float(amut_ext))/amut_noext * 100)
    sum_perc_dec += perc_decrease
    perc_decrease = str(round(perc_decrease,1))
    res1.loc[i] = [tc1, amut_noext, f"{amut_ext} (-{perc_decrease}%)"]
res1
```

```python
sum_perc_dec = 0
res2 = pd.DataFrame(columns=["", "NOEXT", "EXT"])
print("Avg. worst case response time")
for i in range(11):
    tc1 = df_noext.iloc[i]["Testcase"]
    tc2 = df_ext.iloc[i]["Testcase"]
    assert tc1 == tc2
    amut_noext = df_noext.iloc[i]["Avg. worst case response time"].item()
    amut_ext = df_ext.iloc[i]["Avg. worst case response time"].item()
    
    perc_decrease = ((float(amut_noext) - float(amut_ext))/amut_noext * 100)
    sum_perc_dec += perc_decrease
    perc_decrease = str(round(perc_decrease,1))
    res2.loc[i] = [tc1, amut_noext, f"{amut_ext} (-{perc_decrease}%)"]
res2
```

```python
sum_perc_dec = 0
res3 = pd.DataFrame(columns=["", "NOEXT", "EXT"])
print("Avg. avg response time")
for i in range(11):
    tc1 = df_noext.iloc[i]["Testcase"]
    tc2 = df_ext.iloc[i]["Testcase"]
    assert tc1 == tc2
    amut_noext = df_noext.iloc[i]["Avg. avg response time"].item()
    amut_ext = df_ext.iloc[i]["Avg. avg response time"].item()
    
    perc_decrease = ((float(amut_noext) - float(amut_ext))/amut_noext * 100)
    sum_perc_dec += perc_decrease
    perc_decrease = str(round(perc_decrease,1))
    res3.loc[i] = [tc1, amut_noext, f"{amut_ext} (-{perc_decrease}%)"]
res3
```

```python
sum_perc_dec = 0
res4 = pd.DataFrame(columns=["", "NOEXT", "EXT"])
print("Total cost")
for i in range(11):
    tc1 = df_noext.iloc[i]["Testcase"]
    tc2 = df_ext.iloc[i]["Testcase"]
    assert tc1 == tc2
    amut_noext = df_noext.iloc[i]["Total cost"].item()
    amut_ext = df_ext.iloc[i]["Total cost"].item()
    
    perc_decrease = ((float(amut_noext) - float(amut_ext))/amut_noext * 100)
    sum_perc_dec += perc_decrease
    perc_decrease = str(round(perc_decrease,1))
    res4.loc[i] = [tc1, amut_noext, f"{amut_ext} (-{perc_decrease}%)"]
res4
```

```python
df_res = pd.concat([res1,res2.iloc[:,1:],res3.iloc[:,1:], res4.iloc[:,1:]], axis=1)
df_res
```

```python
# Convert to string for LaTeX table
x = df_res.to_string(header=False,
                  index=False,
                  index_names=False).replace("%", "\\%").replace(" (", "_(").split('\n')
x
```
