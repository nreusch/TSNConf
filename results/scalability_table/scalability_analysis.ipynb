```python
import pandas as pd

dfs = {}
df_cp = pd.read_csv(f"best_cp.csv")
df_sa = pd.read_csv(f"best_sa.csv")
```

```python
df_cp.sort_values(by="Testcase")
```

```python
df = df_cp.loc[df_cp["Testcase"] == "giant2"]
df
```

```python
columns = ["Test case", "Method", "# ES", "# SW", "# Streams", "# Recv. Tasks", "# Tasks", "Cost", "T"]
tc_name = ["tiny1", "tiny2", "tiny3", "small1", "small2", "small3", "medium1", "medium2", "medium3", "large1", "large2", "large3", "huge1", "huge2", "huge3", "giant1", "giant2", "giant3"]
df_result = pd.DataFrame([], columns = columns)

i = 0
for name in tc_name:
    # Parse CP results
    df = df_cp.loc[df_cp["Testcase"] == name]
    if not df.empty:
        method = "CP"
        nr_es = int(df["# ES"])
        nr_sw = int(df["# Switches"])
        nr_s = int(df["# Streams"])
        nr_recv_tasks = int(df["# Receiver Tasks"])
        nr_tasks = int(df["# Tasks"])
        cost = int(df["Total cost"])
        t = int(df["Total Runtime (ms)"])
        df_result.loc[i] = [name, method, nr_es, nr_sw, nr_s, nr_recv_tasks, nr_tasks, cost, t]
    else:
        method = "CP"
        df_result.loc[i] = [name, method, "/", "/", "/", "/" , "/" , "/" , "/"]
        
    # Parse SA results
    df = df_sa.loc[df_sa["Testcase"] == name]
    if not df.empty:
        method = "SA"
        nr_es = int(df["# ES"])
        nr_sw = int(df["# Switches"])
        nr_s = int(df["# Streams"])
        nr_recv_tasks = int(df["# Receiver Tasks"])
        nr_tasks = int(df["# Tasks"])
        cost = int(df["Total cost"])
        t = int(df["Total Runtime (ms)"])
        df_result.loc[i+1] = [name, method, nr_es, nr_sw, nr_s, nr_recv_tasks, nr_tasks, cost, t]
    else:
        method = "SA"
        df_result.loc[i+1] = [name, method, "/", "/", "/", "/" , "/" , "/" , "/"]
    i += 2
df_result
```

```python
df_result.to_csv("impact_results.csv")
```
