```python
import pandas as pd

df = pd.read_csv(f"tec2021_edgeapps.csv")
df = df[["Method","Edge Applications used","Avg. edge app latency"]]
df
```

```python
import math
sum_perc_dec = 0
res = pd.DataFrame(columns=["", "NOEXT", "EXT"])
for i in range(7):
    lat_noext = df[df["Edge Applications used"]==f"testcases/ra/RA_example_edge_apps_{i}.flex_network_description"].iloc[1]["Avg. edge app latency"].item()
    lat_ext = df[df["Edge Applications used"]==f"testcases/ra/RA_example_edge_apps_{i}.flex_network_description"].iloc[0]["Avg. edge app latency"].item()
    
    perc_decrease = ((float(lat_noext) - float(lat_ext))/lat_noext * 100)
    sum_perc_dec += perc_decrease
    perc_decrease = str(round(perc_decrease,2))
    res.loc[i] = [f"Edge Apps {i+1}", lat_noext, f"{lat_ext} (-{perc_decrease}%)"]
res
```

```python
res.to_csv("tec2021_edgeapps_results.csv")
```
